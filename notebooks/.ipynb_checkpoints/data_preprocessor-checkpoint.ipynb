```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import data_loader2

# Import the load_data function from the data_loader module
#from data_loader import load_csv

# Load dataset using the function from the data_loader
df = data_loader2.load_csv('/Users/damonnamin/desktop/MLHealth/data/diabetes-dataset.csv')  

# Handling missing values
imputer = SimpleImputer(strategy='mean')
df.iloc[:, :] = imputer.fit_transform(df)

# Dealing with outliers using Z-score
z_scores = (df - df.mean()) / df.std()
df = df[(z_scores > -3).all(axis=1) & (z_scores < 3).all(axis=1)]

# Normalizing data
scaler = StandardScaler()
df.iloc[:, :] = scaler.fit_transform(df)

# Addressing class imbalance
y = df['Outcome']  # Target variable
X = df.drop('Outcome', axis=1)  # Replace 'target_column' with your target column name

y = y.astype(int)
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(y.dtype)
# Use SMOTE to oversample the minority class in the training set
smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)
```
